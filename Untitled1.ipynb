```python
all_data = pd.read_sql_query("SELECT * FROM chart_data INNER JOIN trading_session ON chart_data.session_id = trading_session.id", conn)
all_data['datetime'] = pd.to_datetime(all_data[['date', 'time']].apply(lambda row: ' '.join(row.values.astype(str)), axis=1))
all_data = all_data.iloc[:, [j for j, c in enumerate(all_data.columns) if j not in [1, 6, 7]]]
all_data = all_data.sort_values(['datetime', 'deal_id', 'id'])
all_data = all_data.drop_duplicates(subset=['deal_id', 'trading_type', 'platform_id'], keep='first')
```

```python
import matplotlib.pyplot as plt
sessions = all_data["session_id"].drop_duplicates()
```

```python
def plot(data):
    clrs = ['blue', "purple", "red", "green", "black"]
    i=0
    su=0
    for s in sessions:
        d = data[data["session_id"] == s]
        x = list(range(su, su + len(list(d["price"]))))
        y = list(d["price"])
        plt.plot(x, y, color=clrs[i])
        i += 1
        i %= 5
        su+=len(d)
    plt.show()
    plt.close()
```

```python
from datetime import timedelta

sessions = set(all_data['session_id'].to_list())
to_drop = set()
for session in sessions:
    df = all_data[all_data['session_id'] == session]
    if df['datetime'].iloc[-1] - df['datetime'].iloc[0] > (timedelta(hours=1) if df['trading_type'].iloc[0] == 'monthly' else timedelta(minutes=30)):
        to_drop.add(session)

all_data = all_data[all_data['session_id'].apply(lambda t: t not in to_drop)]
```

```python
all_data
```

```python
for trading_type in ['monthly', "daily"]:
    for platform_id  in [1, 2]:
        series = []
        data = all_data[(all_data["trading_type"] == trading_type) & (all_data["platform_id"] == platform_id)]
        for session_id in list(sorted(data["session_id"])):
            for i in range(60):
                session_data = data[data["session_id"] == session_id]
                serie_size = 30 if trading_type == 'daily' else 60
                
                deals = [0 for _ in range(serie_size)]
                lots = [0 for _ in range(serie_size)]
                for _, row in session_data.iterrows():
                    idx = row['datetime'].minute % serie_size
                    deals[idx] += row['price'] * row['lot_size']
                    lots[idx] += row['lot_size']

                result = [0 for _ in range(serie_size)]
                last_price = session_data['price'].iloc[0]

                for i in range(serie_size):
                    result[i] = last_price if lots[i] == 0 else deals[i] / lots[i]
                    last_price = result[i]
                series.append(result)
```
